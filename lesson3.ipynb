```julia
f(x) = x^2 - 2
f_prime(x) = 2x
f_doubleprime(x) = 2.0

# Function to perform Newton's method for MINIMIZATION
function newton_method(f, f_prime, f_doubleprime, x0, tol, max_iter = 1000)
    x = x0
    for i in 1:max_iter
        # Newton's update: x = x - f'/f''
        step = f_prime(x) / f_doubleprime(x)
        x = x - step
        
        println("Iteration $i: x = $x, f(x) = $(f(x))")
        
        # Stop when the derivative is effectively zero (stationary point)
        if abs(f_prime(x)) < tol
            break
        end
    end
    return x
end

# Running the solver
result_x = newton_method(f, f_prime, f_doubleprime, 1.0, 1e-5)
```

```julia
# Plotting the function
using Plots
x = -2:0.1:2
y = f.(x)
plot(x, y, title="Plot of f(x) = x^2 - 2", xlabel="x", ylabel="f(x)", legend=false)
hline!([0], linestyle=:dash, color=:red)
scatter!([result_x], [f(result_x)], color=:green, label="Minimum found")
```
