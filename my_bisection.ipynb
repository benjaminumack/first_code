```julia
function solve_bisection(f, a, b, tol, max_iter)
    if f(a) * f(b) >= 0
        error("f(a) and f(b) must have opposite signs")
    end

    for iter in 1:max_iter
        c = (a + b) / 2
        if abs(f(c)) < tol || (b - a) / 2 < tol
            return c
        end
        if f(c) * f(a) < 0
            b = c
        else
            a = c
        end
    end
    error("Maximum iterations reached without convergence")
    
end
```

```julia
function solve_bisection(f, a, b, tol, max_iter)
    # 1. If an endpoint is already the root, just return it!
    if f(a) == 0 return a end
    if f(b) == 0 return b end

    # 2. Only error if they are on the SAME side (both + or both -)
    # Using sign() is safer to avoid numerical overflow
    if sign(f(a)) == sign(f(b))
        error("f(a) and f(b) must have opposite signs. f(a)=$(f(a)), f(b)=$(f(b))")
    end

    for iter in 1:max_iter
        c = (a + b) / 2
        
        # 3. Success check
        if abs(f(c)) < tol || (b - a) / 2 < tol
            println("Found root at x = $c after $iter iterations")
            return c
        end

        # 4. Standard Bisection Logic
        if sign(f(c)) == sign(f(a))
            a = c
        else
            b = c
        end
    end
    error("Maximum iterations reached")
end

# Define function and run
f(x) = x^2 - 1
solve_bisection(f, 1.0, 4.0, 1e-6, 1000)
```

```julia
root = solve_bisection(f, a, b, tol, max_iter)
println("Root found at: ", root)
```
